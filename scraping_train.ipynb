```python
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
import requests
import pprint
```

```python
url = "https://qiita.com/advent-calendar/2016/crawler"
r = requests.get(url)
```

```python
soups = BeautifulSoup(r.text,"html.parser")
```

```python
cal = soups.find(class_="table")
names = cal.find_all("div","adventCalendarCalendar_author")
titles = cal.find_all("div",class_="adventCalendarCalendar_comment")
contents = cal.find_all(class_="adventCalendarCalendar_comment")
links = cal.find_all(class_="adventCalendarCalendar_comment")
link = [link.contents[0].get('href').replace("\xa0","") for link in links]
```

```python
name = [name.text.replace("\xa0","") for name in names]
content = [content.text.replace("\xa0","") for content in contents]
dic = {
    "name":name,
    "content":content,
    "link":link
}
dic
```

```python
df = pd.DataFrame(dic)
df.head()
```

```python
for i in link:
    for a in content:
        print(a + "  " + i)
```
